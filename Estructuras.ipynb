# Estructuras de Control 
# Ejemplos de uso de estructuras de control en Python

```python
# Secuencia 
# Esta estructura de control trata de la ejecuacion de una instruccion tras otra instruccion 
num1 = 7 
num2 = 3 

suma = num1 + num2

print(f"La suma es: {suma}")
```

```python
# Iteracion 
# Esta estructura de control trata de la repeticion de un bloque de codigo varias veces
# Ciclo For 

for i in range(5):
    print (f"Interacioón de número {i}") 
```

```python
# Realizar un programa que muestre los primeros N multiplos de un numero M.
n = 10 
m = 7 

for i in range(1, n + 1):
    print(f"7 x {i} =  {m * i}") 
```

```python
# Realizar el factorial de un numero
n = 8 
f = 1 

for i in range(1, n): 
    f *= (i+1)
print(f"{n}! = {f}")
```

```python
# Realizar un programa que calcule la N potencia de un numero M

n = 4 
num = 2
p = 1

for i in range(1, n):
    p = (p*num)
print(f"{n}^ = {p}")
```

```python
# Realizar un programa que calcule el producto de dos numero sin utilizar el signo de multiplicacion 
t = 4
r = 1 

for num in range(4, 40, 4):
    r = (t + num)
print(f"{t}*10 = {r}")
    
```

```python
# Selección Simple 
# Un Programa qeu temrina si la esas ingresada por un ususario corresponde a una 
# persona mayor de edad. Uno es mayor de edad a partir de los 18 años. 

edad = int(input("Ingrada su edad: ")) 

if edad >= 18:
    print("Es mayor de edad ")
else: 
    print("Es menor de edad ")
```

```python
# Program para derterminar si un numero es para o impar ingrasando porteclado 

numero = int(input("Ingrese un numero: "))
if numero == 0:
    print("El nuemro es cero")
elif numero % 2 == 0:
    print("El numero es par ")
else: 
    print("el numero es impar ")
```

```python
# realizar si un nmero es positovo, negativo o cero ingredado porteclado 

numero = int(input("Ingrese un numero: "))
if numero == 0:
    print("El numero es Neutro")
elif numero >= 0: 
    print("El numero positivo")
else: 
    print("El numero es negativo")
```

```python
# Seleccion Múltiple 
# Programa qeu determine el correspondiente dia de la semana, tras 
# leer un nemro por teclado 
numero = int(input("Ingese un número del 1 la 7"))

match numero: 
    case 1: print("Lunes")
    case 2: print("Martes")
    case 3: print("Miercoles")
    case 4: print("Jueves")
    case 5: print("Viernes")
    case 6: print("Sábado")
    case 7: print("Domingo")
    case _: print("No es un dia de la semana")
```
