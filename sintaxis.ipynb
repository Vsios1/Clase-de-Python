# **Operadores en Python** 

```python   
- Operedores de Asignación 
- Operadores Aritméticos
- Operadores de Relación 
- Operadores Lógicos
- Asignacion Complejas 


````python
# Asignación 
print("Asignacion")
num_fav = 8 
num1 = 25 
num2 = 94
print('Numero favorito:', num_fav)

# Operadores Aritméticos
print(num1, "+", num2, "=", num1 + num2) # Suma
print(num1, "-", num2, "=", num1 - num2) # Resta
print(num1, "*", num2, "=", num1 * num2) # Producto
print(num1, "/", num2, "=", num1 / num2) # División Decimal
print(num1, "//", num2, "=", num1 // num2) # División Entera
print(num1, "%", num2, "=", num1 % num2) # Módulo

# Operaciones Relacionales
print("\nOperaciones Relacionales:")
print(f"{num1} > {num2} = {num1 > num2}")  
print(f"{num1} < {num2} = {num1 < num2}")
print(f"{num1} <> {num2} = {num1 != num2}")  
print(f"{num1} >= {num2} = {num1 >= num2}")
print(f"{num1} <= {num2} = {num1 <= num2}")
print(f"{num1} == {num2} = {num1 == num2}")
print(f"{num1} != {num2} = {num1 != num2}")

# Asignacion Complejas: 
num1 += num2 # num1 = num1 + num2
print('El nuevo valor de num1 es:', num1) 

num2 -= num1 # num2 = num2 - num1
print('El nuevo nueor de num2 es:', num2)

num1 *= num2 # num1 = num1 * num2
print('El nuevo valor de num1 es:', num1)

num2 /= num1 # num2 = num2 / num1
print('El nuevo numero de num2 es:', num2)

num1 %= num2 # num1 = num1 % num2 
print('El nuevo muneor de num1 es:', num1)

#Operadores Lógicos
print("\nOperadores Lógicos:")
# AND -> Devuelve True si ambos operandos son True

logico1 = logico2 = False 
logico3 = logico4 = True 

# and (y logico) -> Devuelve True si ambos operandos son True
print(logico1 and logico1)
print(logico1 and logico2)
print(logico1 and logico3)
print(logico1 and logico4)

# or (o logico) -> Devuelve True si al menos exixste un True 
print(logico1 or logico1)
print(logico1 or logico2)
print(logico1 or logico3)
print(logico1 or logico4) 

````
